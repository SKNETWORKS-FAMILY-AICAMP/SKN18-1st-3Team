```python
import pandas as pd
import os

# current_dir = os.path.dirname(__file__)


df = pd.read_excel("202201_202412_전기차등록현황.xlsx")
```

```python
df.head()
```

```python
df['년월'] = pd.to_datetime(df['년월'], format='%Y-%m')
df = df.sort_values('년월').reset_index(drop=True)
df.head()
```

```python
df.columns
```

```python
# df_raw: 등록 대수
region_cols = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종', '경기', '강원', '충북','충남', '전북', '전남', '경북', '경남', '제주', '합계']

df_diff = df.copy() 
df_diff = df[region_cols].diff()
df_rate = df[region_cols].pct_change() * 100

# 컬럼명 정리
df_diff.columns = [f"{col}_증가량" for col in df_diff.columns]
df_rate.columns = [f"{col}_증가율" for col in df_rate.columns]

# 최종 월별 변화 데이터프레임 생성
df_monthly_change = pd.concat([df[['년월']], df_diff, df_rate], axis=1)

# 확인
df_monthly_change.head()
```

```python
df_monthly_change = df_monthly_change.iloc[1:,:]
df_monthly_change.head()
```

```python
df_rate = df_monthly_change.columns.str.replace('_증가율', '', regex=False)
df_rate.head()
```
